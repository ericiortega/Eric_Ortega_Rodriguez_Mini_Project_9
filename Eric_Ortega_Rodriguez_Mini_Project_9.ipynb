```python
import pandas as pd
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Load the CSV
url = 'https://github.com/nogibjj/Eric_Ortega_Rodriguez_Mini_Project_8/raw/refs/heads/main/data/cereal.csv'
cereal_df = pd.read_csv(url)

# Display
print("First few rows of the dataset:")
display(cereal_df.head())

# Simple Data Manipulation:
# 1. Filter cereals with more than 100 calories
high_calorie_cereals = cereal_df[cereal_df['calories'] > 100]
print("\nCereals with more than 100 calories:")
display(high_calorie_cereals)

# 2. Average sugar amount
average_sugar = cereal_df['sugars'].mean()
print(f"\nAverage sugar content of cereals: {average_sugar:.2f} grams")

# 3. Sort cereals by protein amount
sorted_by_protein = cereal_df.sort_values(by='protein', ascending=False)
print("\nCereals sorted by protein content:")
display(sorted_by_protein.head())
```

Visualization of cereals with high calories

```python
plt.figure(figsize=(12, 10))
calories_heatmap = cereal_df[['name', 'calories']].set_index('name')
sns.heatmap(calories_heatmap, annot=True, fmt="g", cmap='YlOrRd', linewidths=0.5, cbar_kws={'label': 'Calories'})
plt.title('Heatmap of Calories in Cereals Dataset', fontsize=16)
plt.xlabel('')
plt.ylabel('Cereal Name', fontsize=12)
plt.xticks(fontsize=12)
plt.yticks(fontsize=10, rotation=0)
plt.tight_layout()
plt.show()
```
